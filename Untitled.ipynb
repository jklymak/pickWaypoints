```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
import cPickle as pickle
import plotTopo
```

```python
# load the stations
fin = open('stations.txt','rb')
sta={}
sta['name']=[]
sta['lat']=np.array([])
sta['lon']=np.array([])

for nn,line in enumerate(fin):
    if nn>2:
        sts=line.split(' ')
        sta['name'].append(sts[0])
        sta['lat']= np.append(sta['lat'], np.array(float(sts[1])))
        sta['lon']= np.append(sta['lon'],-np.array(float(sts[2])))
pickle.dump(sta,open('Stations.pickle','w'))    
```

```python
fig=plt.figure(figsize=(12,12))
map=plotTopo.plotTopo(fig)
for nn in range(len(sta['name'])):
    map.plot(sta['lon'][nn],sta['lat'][nn],'rd',latlon=True)
map.fillcontinents(zorder=100)
```
